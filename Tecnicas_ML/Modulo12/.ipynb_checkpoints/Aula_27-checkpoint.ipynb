# Theory

## Polynomial Regression
- not all relationships are linear
- linear formula (aka first degree) $y = ax + b$
- Second order polynom $y = ax^2 + bx + c$
- Third order polynom $y = ax^3 + bx^2 + cx + d$

## Beware overfitting
- Dont use more degrees than need
- Visualize the data first
- Visualize the fit
- A  high r-squared simply means the curve fits well the trainning data, but may be not a good predictor


# Code

```python
# Import das bibliotecas
%matplotlib inline
from pylab import *
import numpy as np

# Gerando distribuições aleatorias (com seed para ter resultados consisos)
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

# Plot do scatter dos dados
scatter(pageSpeeds, purchaseAmount)
```

```python
# Seta os eixos dos dados
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

# Função numpy que tenta encaixar um polinomio, nesse caso de 4 ordem
p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
import matplotlib.pyplot as plt

# Gera uma sequencia de 0 a 7 com 100 pontos igualmente espaçados
xp = np.linspace(0, 7, 100)
# Plota os pontos e a aproximação polinomial
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print(r2)
```

# Graficos para o relatorio

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados de exemplo
np.random.seed(0)
x = np.linspace(-10, 10, 100)
y = 0.5 * x**2 + 3 * x + 5 + np.random.normal(0, 10, size=x.shape)  # Dados com ruído

# Regressão linear
coef_linear = np.polyfit(x, y, 1)
poly_linear = np.poly1d(coef_linear)
y_linear_fit = poly_linear(x)

# Regressão polinomial (grau 2)
coef_poly = np.polyfit(x, y, 2)
poly_poly = np.poly1d(coef_poly)
y_poly_fit = poly_poly(x)

# Criando o plot lado a lado
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plotando regressão linear
axs[0].scatter(x, y, color='gray', label='Dados')
axs[0].plot(x, y_linear_fit, color='red', label='Linear')
axs[0].set_title('Regressão Linear')
axs[0].legend()

# Plotando regressão polinomial
axs[1].scatter(x, y, color='gray', label='Dados')
axs[1].plot(x, y_poly_fit, color='blue', label='Polinomial (grau 2)')
axs[1].set_title('Regressão Polinomial (Grau 2)')
axs[1].legend()

plt.show()
```
