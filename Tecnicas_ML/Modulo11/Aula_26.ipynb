# Linear Regression
- Fit a line to a data set of observations

## How it works
- Usually "least squares"
- Minimizes the squared-error between each point and the line
- Least squares minimizes the sum of squared errors
- This is the same as maximizing the likelihood of the observed data
- This is sometimes called maximum likelihood estimation

## There is more than 1 way to do it
- Gradient Descent
- Iterates to find the line that best follows the contours defined by data
- Can make sens when dealing with 3d data

## Measuring error with r-squared
- How do we measure how well our line fits our data?
- R-squared (aka coefficient of determination) measures

## Interpreting r-squared
- rangest to 0 to 1
- o bad 1 good

```python
# Import das bibliotecas
%matplotlib inline
import numpy as np
from pylab import *

# Geração de dados aleatorios, sendo PA dependente de PS
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

# plot dos dados
scatter(pageSpeeds, purchaseAmount)
```

```python
# import da scipy, parte de estatistica
from scipy import stats

# Faz a regressão linear, utilizando a scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
r_value ** 2
```

```python
# Import da matplotlib
import matplotlib.pyplot as plt

# Gera os dados para a linha da regressão
def predict(x):
    return slope * x + intercept
fitLine = predict(pageSpeeds)

# Plota ambos o scatter e a linha
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```

```python
def test(mean, var):
    
    pageSpeeds = np.random.normal(mean, var, 1000)
    purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
    slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
    print(f'r-squared: {r_value**2}')
```

```python
# Controle
test(3.0, 1.0)
```

```python
# Media == variancia
mean = 3.0
var = 3.0
test(mean, var)
```

```python
# Media < variancia
mean = 3.0
var = 6.0
test(mean, var)
```

```python
# Media > variancia
mean = 3.0
var = 0.1
test(mean, var)
```

```python
# Media > variancia
mean = 0.1
var = 3.0
test(mean, var)
```

```python
# Media <<< variancia
mean = 3.0
var = 10e10
test(mean, var)
```

```python
# Media > variancia
mean = 1000
var = 1.0
test(mean, var)
```

Para esse caso de regressão linear, pela forma que a "linha" é formada, quanto mais esparços os dados mais preciso o modelo.
