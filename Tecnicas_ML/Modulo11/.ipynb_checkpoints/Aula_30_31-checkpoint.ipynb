# What is Machine Learning?
 Algorithms that can learn from observational data, and can make predictions based on it



## Unsupervised Learning
The model is not given "answers" to learn, it must make sense of the data just by observing.

- Example: group (cluster) some object into 2 different sets. But the model dont have the info on what the right or wrong is.

### Why use it?
When whe dont know what relationships we are looking for, and "let the data talk"

Example:
- Clustering users on a dating site based on their information and behavior. Perhaps you'll find there are groups of people that match but dont conform to stereotypes.
-  Cluster movies based on their properties
- Analyze the text of product descriptions.

## Supervide Learning
In supervised learning, the data algorithm learn from correct and wrong answers. The model then is used to predict from new, unknown values.
- Example: Train a model from predicting car prices based on car attributes using historical sales data. That model can the predict the optimal price for new cars that haven't ben sold before.

### Evaluating Supervised Learning
- If you have a set of training data that includes the value you're trying to predict - you dont have to guess if the result model is good or not.
- If you have enought training data, you can split it in 2 parts, training and test sets.
- Then when the model is trained, we can use r-squared or some other metric to evaluate the accuracy, by using the test set.

### Train/Test in practice
- Need to ensure both sets are large enough to conatin representatives of all variations and outliers in the data.
- The data sets must be selected randomly.
- Train/test is a great way to guard against overfitting.
- But that is still not infallible.
- K-fold can help in that.

# Code

```python
# Import das bibliotecas
%matplotlib inline
import numpy as np
from pylab import *

# Gera os dados aleatorios
np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds

# Plota os dados
scatter(pageSpeeds, purchaseAmount)
```

```python
# Separa em sets de treino e teste
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
# Plota os dados de treino
scatter(trainX, trainY)
```

```python
# Plota os dados de teste
scatter(testX, testY)
```

```python
# Transforma os dados em np.array
x = np.array(trainX)
y = np.array(trainY)

# Realiza a regressão polinomial (polinomio de 8 graus)
p4 = np.poly1d(np.polyfit(x, y, 8))
```

```python
# Pota a aproximação gerada pela regressão sobreposta pelos pontos de treino
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# Pota a aproximação gerada pela regressão sobreposta pelos pontos de teste
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# Metrica para os dados de teste
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2)
```

```python
# Metrica para os dados de treino
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX)))

print(r2)
```

## Activity

```python
# Gera um novo conjunto de dados

np.random.seed(69)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

# Separa em sets de treino e teste
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]

x = np.array(trainX)
y = np.array(trainY)
```

```python
# Brute forcing por varios graus de polinomios
pol = []
r_values = []
for i in range(1, 18):
  pol.append(np.poly1d(np.polyfit(x, y, i)))
  r_values.append(r2_score(np.array(trainY), pol[i-1](np.array(trainX))))
```

```python
index = 1
for i in r_values:
  print(f'r2 polinomio de grau {index}: {i}')
  index+=1
```

```python
# Plot do polinomio de grau 17 com os pontos de treino
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, pol[16](xp), c='r')
plt.show()
```

```python
# Plot do polinomio de grau 17 com os pontos de teste
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, pol[16](xp), c='r')
plt.show()
```
