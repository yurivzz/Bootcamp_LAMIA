# Aula 14 - Média, Mediana e Moda

## Média

Soma de todas as amostras pelo numero de amostras.

```python
# Exemplo simples do calculo da média
amostras = [0, 2, 3, 2, 1, 0, 0, 2, 0]
numero_amostras = len(amostras)

media = sum(amostras)/numero_amostras

print(f'A média é:{media:.2f}')
```

## Mediana

A mediana é o valor central do conjunto de dados ordenado.
Temos 2 casos possiveis de mediana:
  - Número impar de dado: a mediana é o valor exatamente no meio do dataset
  - Número par de dados: a mediana sera a média entre os 2 valores do meio

A mediana é menos suscetível a ser "deslocada" por valores muito fora da curva do dataset (por exemplo a renda média pareceria maior por conta de bilionários que concentram montantes absurdo em relação ao resto da população)

```python
# Exemplo simples do calculo da mediana (apenas para numero de amostras impares)
amostras.sort()
print(f'O conjunto de dados ordenado é: {amostras}')
print(f'A mediana será: {amostras[numero_amostras // 2]}')
```

## Moda

O valor que mais se repete no dataset, não é relevante em dados contínuos.

```python
# Exemplo simples do calculo da moda, utilizando o metodo Counter
from collections import Counter

print(f'Contagem das amostras:{Counter(amostras)}')
print('O elemento que mais se repete é o 0')
```

# Aula 16 - Desvio Padrão e Variância

## Variância

Notada como $\sigma^2$, é a média dos quadrados de cada elemento do dataset subtraido pela média do proprio dataset.

Definida matematicamente como:

$
    \sigma^2 = E[[X - E[X]]^2]
$

Sendo $E$ o valor esperado.

```python
# Calculo da variancia, tiramos a média dos dados, subtraimos dos dados a média, elevamos ao quadrado e por fim tiramos a média novamente, resultando na variancia
import numpy as np
data = np.array([1, 4, 5, 4, 8])
mean = np.mean(data)
dif_from_mean = data - mean
sqr_dif_from_mean = dif_from_mean ** 2
variance = np.mean(sqr_dif_from_mean)
variance
```

## Desvio Padrão

Simplesmente a raiz quadrada da variacia, notado por σ

σ = $ \sqrt{σ^2} $

```python
# Simplesmente a raiz quadrada da variancia
desvio_padrao = np.sqrt(variance)
desvio_padrao
```

## População vs. Amostras

A diferença entre a variância de população e a de amostras é que, se tivermos N amostras, então apenas dividimos por N-1 (quando calculada a média "final" da variância).

Em matematriques:

$
  \sigma^2 = \frac{\Sigma(X-\mu)^2}{N}
$

$
  S^2 = \frac{\Sigma(X-M)^2}{N-1}
$

# Aula 17 - Density Function; Probability Mass Function

São respectivamente, o modelo continuo da função e o discreto, como os plots a seguir mostram.

## Exemplo: Função densidade de probabilidade (PDF) de uma distribuição normal (gaussiana)

```python
# Codigo para plotar a PDF de uma distribuição gaussiana
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0
sigma = 1

# Gera os pontos de -4 a 4
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Gera a PDF
pdf = norm.pdf(x, mu, sigma)

# Plota a PDF
plt.plot(x, pdf, label='Distribuição Normal')
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Normal')
plt.legend()
plt.grid(True)
plt.show()
```

## Exemplo: Função de massa de probabilidade (PMF) de uma distribuição normal (gaussiana)


```python
# Simplesmente um histograma da distribuição(PMF) sobreposto com a PDF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu = 0
sigma = 1

#Gera o histograma da distribuição
samples = np.random.normal(mu, sigma, size=1000)

plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Basicamente o mesmo codigo da celula acima
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k', linewidth=2, label='Distribuição Normal')

plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')
plt.title('Aproximação Discreta da Distribuição Normal')
plt.legend()
plt.grid(True)
plt.show()
```

## Principais diferenças:

- Tipo de Variável: A PDF é para variáveis contínuas, enquanto a PMF é para variáveis discretas.
- Representação de Probabilidades: A PDF usa áreas sob a curva para representar probabilidades de intervalos, enquanto a PMF atribui diretamente probabilidades a valores específicos.
- Integração vs. Soma: Na PDF, as probabilidades são encontradas integrando sobre um intervalo, enquanto na PMF, as probabilidades são encontradas somando valores específicos.

# Aula 22 - Covariance and Correlation


## Covariancia

A medida estatistica que diz como 2 variaveis variam em relação as suas médias

Calculada da seguinte forma:

$
cov(X,Y) = E[(X-E[X])(Y-E[Y])]
$

Sendo E a esperança ou o valor esperado (média)

```python

%matplotlib inline

import numpy as np
from pylab import *

# subtrai a média de cada elemento de um array
def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

# Calcula a covariancia (produto escalar de 2 vetores retirados se suas medias)
def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

```python
# outro exemplo de covariancia, porem agora, os dados são "normalizados" (o que diminuiu a covariancia)

purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

## Correlação

Medida que diz o quão 2 variaveis estão linearmente correlacionadas, definida matematicamente como:

$
\rho_{XY} = \frac{cov(X,Y)}{\sigma_X \sigma_Y}
$

Sendo $cov()$ a covariancia e $\sigma$ o desvio padrão
É basicamente a covariancia normalizada, sendo que $\rho_{XY} = -1$ indica que é uma correlação inversa, $\rho_{XY} = 0$ não existe correlação e $\rho_{XY} = 1$ uma correlação perfeita.


Não necessariamente implica casualidade.

```python
# Calculo da correlação, basicamente a covariancia dividida pelo desvio padrao de x e y
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy

correlation(pageSpeeds, purchaseAmount)
```

```python
# método que já existe na numpy, porem gera uma matiz (imagino que seja util para calculos futuros)
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
# Grafico de uma correlação inversa aka -1
purchaseAmount = 100 - pageSpeeds * 3

scatter(pageSpeeds, purchaseAmount)

correlation (pageSpeeds, purchaseAmount)
```

```python
np.cov(pageSpeeds, purchaseAmount)
```

# Aula 23/24 - Conditional Probability

Dados 2 eventos correlacionados A e B, temos que:
$
  P(B|A) = \frac{P(A,B)}{P(A)}
$

Sendo:
  - P(A) a probabilidade de A acontecer
  - P(A,B) a probabilidade de A e B acontecerem
  - P(B|A) a probabilidade de B sendo que A já aconteceu.

```python
# Um simulador de eventos simples, caso o numero aleatorio seja maior que a probabilidade de compra, se marca uma compra para aquela idade
from numpy import random
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0
for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = float(ageDecade) / 100.0
    totals[ageDecade] += 1
    if (random.random() < purchaseProbability):
        totalPurchases += 1
        purchases[ageDecade] += 1
```

```python
# Mostra os resultados da simulação
totals
```

```python
purchases
```

```python
totalPurchases
```

```python
# Calculo da probilidade de comprar, tendo 30 anos
PEF = float(purchases[30]) / float(totals[30])
print('P(purchase | 30s): ' + str(PEF))
```

```python
# calculo da probabilidade de se ter 30 anos
PF = float(totals[30]) / 100000.0
print("P(30's): " +  str(PF))
```

```python
# Probabilidade de compra
PE = float(totalPurchases) / 100000.0
print("P(Purchase):" + str(PE))
```

```python
# Probabilidade de comprar e de ter 30 anos
print("P(30's, Purchase)" + str(float(purchases[30]) / 100000.0))
```

```python
# Probabilidade de comprar multiplicada pela de ter 30 anos
print("P(30's)P(Purchase)" + str(PE * PF))
```

```python
# Simplesmente assumir uma probabilidade de compra, retirando a relação entre idade e a compra
from numpy import random
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0
for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = 0.1
    totals[ageDecade] += 1
    if (random.random() < purchaseProbability):
        totalPurchases += 1
        purchases[ageDecade] += 1


PEF = float(purchases[30]) / float(totals[30])
print("P(purchase | 30s): " + str(PEF))

PE = float(totalPurchases) / 100000.0
print("P(Purchase):" + str(PE))
```

# Aula 25 - Bayes' Theorem


O Teorema de Bayes é um dos princípios fundamentais da probabilidade e da estatística, utilizado para atualizar a probabilidade de uma hipótese com base em novas evidências ou informações, é definido por:

$
  P(A|B) = \frac{P(A)P(B|A)}{P(B)}
$

Onde:
  - P(A|B) é a probabilidade de A ocorrer sendo que B ocorreu
  - P(B|A) é a probabilidade de B ocorrer sendo que A ocorreu
  - P(A) a probabilidade de A ocorrer
  - P(B) a probabilidade de B ocorrer
